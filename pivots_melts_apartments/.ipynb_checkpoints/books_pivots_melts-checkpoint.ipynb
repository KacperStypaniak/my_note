```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

books = pd.read_csv('books.csv',skiprows = [3349,4703,5878,8980])

books = pd.read_csv('books.csv',on_bad_lines = 'skip',engine = 'python')

books
# avg_lg_rt = books.groupby(['language_code'])[['average_rating','ratings_count']].mean().reset_index()

# # avg_lg_rt.pivot(index = lan)
# avg_lg_rt





# użyj pivot_table by wyciągnąć ,maksymalny rating dla języka i maksymalną liczbę recenzji

max_lg_rt = books.pivot_table(index = 'language_code', values = ['average_rating','ratings_count'], aggfunc = 'max')
max_lg_rt.columns = ['max_rating', 'max_ratings_count']

max_lg_rt


# użyj pivot_table by wyciągnąć ,średni rating dla języka i sumę liczb recenzji

avg_max_lg_rt = books.pivot_table(index = 'language_code'
                                  , values = ['average_rating','ratings_count']
                                  , aggfunc = {'average_rating': 'mean', 'ratings_count': 'sum'})
avg_max_lg_rt.columns = ['avg_rating', 'sum_ratings_count']

avg_max_lg_rt



# dwie powyższe rzeczy można również zrobić groupby? (groupby pokazuje od razu, nie trzeba zmieniać obiektu, ani tworzyć nowego )

books.groupby('language_code').agg(avg_rating = ('average_rating','mean'), sum_ratings_count = ('ratings_count','sum'))
```

```python
# melting


books_melted = books.melt(id_vars = ['bookID','title','authors'], )

hp_books = books_melted[books_melted['title'].str.startswith('Harry Potter')].sort_values('bookID', ascending = True)

# hp_books[hp_books['variable']=='publication_date']

hp_books
```

```python
# utworzyć słownik wszystkich 'variable'


Key = 0
slownik = {}

for zmienna in books_melted['variable'].unique():
    slownik[Key] = zmienna
    Key = Key + 1


slownik.keys()
slownik.values()

pd.DataFrame.from_dict(slownik, orient = 'index', columns = ['Name'])

# slownik = pd.DataFrame(index = slownik.keys(), slownik.values())
# slownik
```
